```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)

normal_data = np.random.normal(loc=50, scale=10, size=1000)
nonnormal_data = np.random.exponential(scale=10, size=1000)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
count, bins, _ = plt.hist(normal_data, bins=30, density=True, color='skyblue', edgecolor='black')
mu, sigma = np.mean(normal_data), np.std(normal_data)
plt.plot(bins, norm.pdf(bins, mu, sigma), 'r', linewidth=2)
plt.title('Normal Distribution\n(Parametric tests valid)')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(1,2,2)
count, bins, _ = plt.hist(nonnormal_data, bins=30, density=True, color='salmon', edgecolor='black')
mu, sigma = np.mean(nonnormal_data), np.std(nonnormal_data)
plt.plot(bins, norm.pdf(bins, mu, sigma), 'r', linewidth=2)
plt.title('Non-Normal Distribution\n(Use Non-Parametric tests)')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
